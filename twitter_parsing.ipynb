```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tweets_data_path = 'twitter_data_2.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```

```python
print(len(tweets_data))
```

```python
tweets_data[10]
```

```python
myMap = {"kn": "Kannada", "hi": "Hindi", "mr": "Marathi", "ur": "Urdu", "gu":"Gujarati", "ta": "Tamil", "ne": "Nepali", "bn": "Bengali", "ml": "Malayalam", "or": "Oriya", "ps": "Pashto", "pa": "Punjabi", "sd": "Sindhi", "si": "Sinhala", "te": "Telugu"}
```

```python
tweets = pd.DataFrame()
tweets['text'] = list(map(lambda tweet: tweet['text'], tweets_data))
tweets['lang'] = list(map(lambda tweet: tweet['lang'], tweets_data))
tweets['language'] = list(map(lambda tweet: myMap[tweet], tweets['lang']))
tweets['country'] = list(map(lambda tweet: tweet['place']['country'] if tweet['place'] != None else None, tweets_data))
tweets['location'] = list(map(lambda tweet: tweet['user']['location'] if tweet['user'] != None else None, tweets_data))
```

```python
plt.figure(figsize=(20,8))
sns.set(style="darkgrid")
ax = sns.countplot(x=tweets['language'],data=tweets)
ax.set_title('Distribution of Tweets by Language')
ax.set_xlabel('Languages')
```

```python
freq = tweets["language"].value_counts()
```

```python
langs = ["Hindi", "Urdu", "Tamil", "Kannada", "Bengali", "Malayalam", "Marathi", "Nepali", "Oriya", "Pashto", "Punjabi", "Sindhi", "Sinhala", "Gujarati"]
```

```python
pops = [339000000, 15100000, 68900000, 43500000, 130000000, 34800000, 83000000, 12300000, 34300000, 20850900, 32600670, 24615550, 15287880, 56408970]
```

```python
egids_vals = [1, 1, 2, 2, 1, 2, 2, 1, 2, 4, 2, 2, 1, 2]
```

```python
egids = {'lang': langs, 'pop': pops, 'egids': egids_vals}
```

```python
egids_df = pd.DataFrame(data=egids)
```

```python
freq.to_frame()
```

```python
egids_df = egids_df.set_index('lang').join(freq)
egids_df = egids_df.reset_index()
egids_df = egids_df.rename(columns={"language": "freq_of_tweets"})
```

```python
value=(egids_df['pop']>50000000)
egids_df['color']= np.where( value==True , "above 50 million", "below 50 million")
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
ax = sns.scatterplot(x=egids_df['lang'], y=egids_df["pop"], data=egids_df, s=400, hue=egids_df["color"])
#ax = sns.relplot(x=egids_df['egids'], y=egids_df["pop"], data=egids_df)
_ = ax.set_title('Language Population vs. EGIDS Rank')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
ax = sns.barplot(x=egids_df['lang'], y=egids_df["pop"], data=egids_df)
_ = ax.set_title('Languages by Population')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
ax = sns.scatterplot(x=egids_df['lang'], y=egids_df["egids"], data=egids_df, s=400, hue=egids_df["color"])
#ax = sns.relplot(x=egids_df['egids'], y=egids_df["pop"], data=egids_df)
_ = ax.set_title('Language Population vs. EGIDS Rank')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
ax = sns.scatterplot(x=egids_df['lang'], y=egids_df["freq_of_tweets"], data=egids_df, s=400, hue=egids_df["color"])
_ = ax.set_title('Language Population vs. Number of Tweets')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
_ =sns.scatterplot(y=egids_df["freq_of_tweets"], x=egids_df["egids"], data=egids_df, s=400, hue=egids_df["color"]);
_ = ax.set_title('EGIDS Rank vs. Number of Tweets')
```

```python
plt.figure(figsize=(20,10))
sns.set(style="darkgrid")
_ =sns.scatterplot(y=egids_df["pop"], x=egids_df["egids"], data=egids_df, s=400, hue=egids_df["color"]);
_ = ax.set_title('EGIDS Rank vs. Population')
```
